```python
import sys
sys.path.append("src")

import numpy as np
from collections import Counter
import quantum_algorithms as qa # customers package 

%matplotlib inline
```

```python
# Parameters
p = 0.1
n = 9
target_qubit = 4

# =====================================================
# Original Deutsch–Jozsa (no error)
# =====================================================

const = qa.deutsch_jozsa.deutsch_jozsa(
    n,
    qa.deutsch_jozsa.f_constant_0
)

bal = qa.deutsch_jozsa.deutsch_jozsa(
    n,
    qa.deutsch_jozsa.f_balanced_parity
)

probs_c = qa.deutsch_jozsa.measure_probs_first_n(const, n)
probs_b = qa.deutsch_jozsa.measure_probs_first_n(bal, n)

print("Original DJA")
print("P(0...0) constant :", probs_c[0])
print("P(0...0) balanced :", probs_b[0])
print()

# =====================================================
# E1 : before first Hadamards
# =====================================================

rho_c1 = qa.depolarizing.deutsch_jozsa_depolarizing1(
    n,
    qa.deutsch_jozsa.f_constant_0,
    p,
    target_qubit
)

rho_b1 = qa.depolarizing.deutsch_jozsa_depolarizing1(
    n,
    qa.deutsch_jozsa.f_balanced_parity,
    p,
    target_qubit
)

probs_c1 = qa.depolarizing.measure_probs_first_n_rho(rho_c1, n)
probs_b1 = qa.depolarizing.measure_probs_first_n_rho(rho_b1, n)

print("DJA with E1 (before H)")
print("P(0...0) constant :", probs_c1[0])
print("P(0...0) balanced :", probs_b1[0])
print()

# =====================================================
# E2 : after first Hadamards
# =====================================================

rho_c2 = qa.depolarizing.deutsch_jozsa_depolarizing2(
    n,
    qa.deutsch_jozsa.f_constant_0,
    p,
    target_qubit
)

rho_b2 = qa.depolarizing.deutsch_jozsa_depolarizing2(
    n,
    qa.deutsch_jozsa.f_balanced_parity,
    p,
    target_qubit
)

probs_c2 = qa.depolarizing.measure_probs_first_n_rho(rho_c2, n)
probs_b2 = qa.depolarizing.measure_probs_first_n_rho(rho_b2, n)

print("DJA with E2 (after first H)")
print("P(0...0) constant :", probs_c2[0])
print("P(0...0) balanced :", probs_b2[0])
print()

# =====================================================
# E3 : after oracle
# =====================================================

rho_c3 = qa.depolarizing.deutsch_jozsa_depolarizing3(
    n,
    qa.deutsch_jozsa.f_constant_0,
    p,
    target_qubit
)

rho_b3 = qa.depolarizing.deutsch_jozsa_depolarizing3(
    n,
    qa.deutsch_jozsa.f_balanced_parity,
    p,
    target_qubit
)

probs_c3 = qa.depolarizing.measure_probs_first_n_rho(rho_c3, n)
probs_b3 = qa.depolarizing.measure_probs_first_n_rho(rho_b3, n)

print("DJA with E3 (after oracle)")
print("P(0...0) constant :", probs_c3[0])
print("P(0...0) balanced :", probs_b3[0])
print()

# =====================================================
# E4 : after final Hadamards
# =====================================================

rho_c4 = qa.depolarizing.deutsch_jozsa_depolarizing4(
    n,
    qa.deutsch_jozsa.f_constant_0,
    p,
    target_qubit
)

rho_b4 = qa.depolarizing.deutsch_jozsa_depolarizing4(
    n,
    qa.deutsch_jozsa.f_balanced_parity,
    p,
    target_qubit
)

probs_c4 = qa.depolarizing.measure_probs_first_n_rho(rho_c4, n)
probs_b4 = qa.depolarizing.measure_probs_first_n_rho(rho_b4, n)

print("DJA with E4 (after final H)")
print("P(0...0) constant :", probs_c4[0])
print("P(0...0) balanced :", probs_b4[0])
```
